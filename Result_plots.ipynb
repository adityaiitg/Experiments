```python
# importing the required module
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
types =  ["tf.compat.v1.lite.experimental.nn.TFLiteLSTMCell","tf.contrib.cudnn_rnn.CudnnLSTM","tf.contrib.rnn.LSTMBlockCell"]
metric = ["acc","loss","time"]
```

```python
acc  = {}
for type in types:
  acc[type] = pd.read_csv(type+"_acc"+".csv")
loss = {}
for type in types:
  loss[type] = pd.read_csv(type+"_loss"+".csv")
time_taken = {}
for type in types:
  time_taken[type] = np.genfromtxt(type+"_time"+".csv", delimiter=",")
  print(np.genfromtxt(type+"_time"+".csv", delimiter=","))
```

```python
# Plotting Accuracy
for type in types:
  plt.plot(acc[type],label = type)
plt.xlabel('Steps')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accuracy Gain over Steps')
plt.show()
```

```python
# Plotting LOss
for type in types:
  plt.plot(loss[type],label = type)
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss vs steps')
plt.show()
```

```python
time_taken_by_types = np.array([])
for type in types:
  time_taken_by_types = np.append(time_taken_by_types,time_taken[type])
```

```python
type_half_name = [t.split('.')[-1] for t in types]
```

```python
plt.scatter(type_half_name,time_taken_by_types)
plt.ylabel('Time in sec')
plt.title('Time comparison')
plt.ylim(0,120)
plt.show()
```
